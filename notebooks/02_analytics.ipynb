```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../db/sales.db')
```

```python
orders = pd.read_sql("SELECT * FROM orders", conn)
orders.head()
```

```python
# Total Revenue
orders['Revenue'].sum()
```

```python
orders.groupby('Region')['Revenue'].sum()
```

```python
orders['Order Date'] = pd.to_datetime(orders['Order Date'])

orders.groupby(orders['Order Date'].dt.to_period('M'))['Revenue'].sum()
```

```python
orders.groupby('Product ID')['Revenue'].sum().sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt

orders.groupby('Region')['Revenue'].sum().plot(kind='bar')
plt.title("Revenue by Region")
plt.show()
```
