```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../db/sales.db')
orders = pd.read_sql("SELECT * FROM orders", conn)
```

```python
total_revenue = orders['Revenue'].sum()
total_orders = len(orders)

print(total_revenue, total_orders)
```

```python
import matplotlib.pyplot as plt
orders.groupby('Order Date')['Revenue'].sum().plot()
plt.title("Revenue Trend")
plt.show()
```

```python
orders.groupby('Product ID')['Revenue'].sum().nlargest(5).plot(kind='bar')
plt.title("Top Products")
plt.show()
```
